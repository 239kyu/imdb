```python
r_list = list['release_url'].tolist()
t_list = list['title_url'].tolist()
```

```python
#타이틀 상세페이지 정보 크롤링


data2 = []

for t in tqdm(t_list, desc='Processing'):

    #url 지정
    url = t

    #헤더설정 및 호출

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    response = requests.get(url, headers=headers)

    html = response.text
    soup3 = BeautifulSoup(html, 'html.parser')

    #타이틀제목
    name = soup3.find('span', {'data-testid': 'hero__primary-text'})
    names = name.text if rating else 'null'


    #평점
    rating = soup3.find('span', {'class': 'sc-bde20123-1 cMEQkK'})
    ratings = rating.text if rating else 'null'

    #출시연도와 러닝타임
    t_box = soup3.find('div', {'class': 'sc-69e49b85-0 jqlHBQ'})
    dat = t_box.find('ul', {'class': 'ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt'}).text if rating else 'null'

    #장르
    genres_tag = soup3.find('div', {'data-testid': 'genres'})
    genres = genres_tag.text if genres_tag else 'null'

    #언어
    language_tag = soup3.find('li', {'data-testid': 'title-details-languages'})
    language = language_tag.text if language_tag else 'null'

    #데이터 주어담기
    data = {'title_name': names,
            'title_url' : url,
            'rating': ratings,
            'releaseyear_runningtime': dat,
            'genres': genres,
            'language': language}

    data2.append(data)

    #time.sleep(random.uniform(1.2))

df2 = pd.DataFrame(data2)

#엑셀저장
df2.to_excel('imdb_title_20210701_20210931.xlsx')
```

```python
#릴리즈 정보 페이지 크롤링

data1 = []

for r in tqdm(r_list, desc='Processing'):

    #URL 지정
    url = r

    #헤더 설정 및 호출

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    response = requests.get(url, headers=headers)

    html = response.text
    soup2 = BeautifulSoup(html, 'html.parser')

    #타이틀 제목
    names = soup2.find_all('h2', {'class': 'sc-a885edd8-9 dcErWY'})
    namess = [name.text for name in names]

    #링크
    link_tag = soup2.find('link', {'rel': 'canonical'})
    r_url = link_tag.get('href')


    #영역설정
    r_div = soup2.find('div', {'data-testid': 'sub-section-releases'})

    #출시국가
    a_tags = r_div.find_all('a', {'class': 'ipc-metadata-list-item__label ipc-metadata-list-item__label--link'})
    r_countries = [a_tag.text for a_tag in a_tags]

    #출시 연월
    span_tags = r_div.find_all('span', {'class': 'ipc-metadata-list-item__list-content-item'})
    r_date = [span_tag.text for span_tag in span_tags]

    #데이터 주어담기
    data = {'title_name': namess * len(r_countries),
            'release_url': r_url * len(r_countries),
            'release_countries': r_countries,
            'release_date': r_date}

    data1.append(data)

df = pd.DataFrame(data1)

#엑셀출력
df.to_excel('imdb_release_20210701_20210931.xlsx')
```
