```python
##설정영역

#데이터프레임 생성
result_df = pd.DataFrame()

#반복문 시작
print('옵션설정 완료, 셀레니움 동작')
for r, d, c in tqdm(zip(pro_url_test, details_url_test, company_url_test), desc='Processing'):
    
    options = webdriver.ChromeOptions()  #옵션할당
    #로그인 쿠키 유지를 위해 셀레니움이 사용할 크롬 프로필 지정
    options.add_argument(r'user-data-dir=C:\Users\kctiuser05\AppData\Local\Google\Chrome\User Data\Profile 3')
    #차단 회피를 위한 헤더 정보 변경
    options.add_argument(f'user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                     f'Chrome/96.0.4664.93 Safari/537.36')
    path = 'chromedriver.exe' #드라이버 경로 지정
    #options = webdriver.ChromeOptions()  #옵션할당
         
    driver = webdriver.Chrome(path, options=options) #드라이버 호출
    wait = WebDriverWait(driver, 3) # 5초 대기를 wait에 할당
    driver.implicitly_wait(2) #3초 대기
                  
    driver.get(r) #url호출  
          
    # 500 픽셀만큼 아래로 스크롤
    driver.execute_script("window.scrollTo(0, 500);")
    # 시간을 주어 스크롤이 완료되기를 기다림
    time.sleep(0.5)
    
    #스타미터 소팅 버튼 클릭 하기 위한 영역
    
    try:
        sortings = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'sort_column_heading')))
        starmeter = sortings[1]
        starmeter.click()
    
    except:
        cast1_data = 'null'
        cast2_data = 'null'
        cast3_data = 'null'
        cast4_data = 'null'
        cast5_data = 'null'       
    
    #try:
    #    starmeter = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_cast_sortable_table"]/tbody/tr[1]/th[2]/span/span/span[1]')))
    #    starmeter.click()    
        #time.sleep(0.3)
        
    #except Exception as e:
    #    print("starmeter_sort에서 에러발생 다음 반복문으로 진행")
    #    driver.quit()
    #    continue
        
    ##스타미터순 출연자 이름과 스타미터 크롤링 영역
    #cast1 영역
    try:
        cast1 = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_cast_sortable_table"]/tr[2]/td[2]/div')))
        cast1_data = cast1.text #데이터 줍기
    except:
        cast1_data = 'null'
        
    #cast2 영역
    try:
        cast2 = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_cast_sortable_table"]/tr[3]/td[2]/div')))
        cast2_data = cast2.text  #데이터 줍기
    except:
        cast2_data = 'null'
    
    #cast3 영역
    try:
        cast3 = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_cast_sortable_table"]/tr[4]/td[2]/div')))
        cast3_data = cast3.text  #데이터 줍기
    except:
        cast3_data = 'null'
        
    #cast4 영역
    try:
        cast4 = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_cast_sortable_table"]/tr[5]/td[2]/div')))
        cast4_data = cast4.text  #데이터 줍기
    except:
        cast4_data = 'null'
        
    #cast5 영역
    try:
        cast5 = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_cast_sortable_table"]/tr[6]/td[2]/div')))
        cast5_data = cast4.text  #데이터 줍기
    except:
        cast5_data = 'null'
            
    #제작사 정보 크롤링 영역
    try:
        #제작사 정보 영역 할당
        production = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'a-fixed-right-grid-inner')))
                
    except:
        company_data = 'failed'
        distributor_data = 'failed'
        rating_data = 'failed'
        print('프로덕션 특정 에러')
        
    try:                
        #제작사
        company = production[2] #특정된 요소 중 2번째 선택
        company_data = company.text #데이터 줍기
    except:
        company_data = 'null'
        
    try:
        #배급사
        distributor = production[3] #특정된 요소 중 2번째 선택
        distributor_data = distributor.text #데이터 줍기        
    except:
        distributor_data = 'null'
        
    try:
        #평점
        #rating = production[4] #특정된 요소 중 2번째
        rating = wait.until(EC.presence_of_element_located((By.XPATH, ' //*[@id="rating_breakdown"]/div[1]/div/div/div[1]/div/div/div[1]/span[2]')))
        rating_data = rating.text #데이터 줍기
    except:
        rating_data = 'null'
                
    ##타이틀 영역
    try:
        #타이틀
        title_heading = driver.find_element(By.XPATH, '//*[@id="title_heading"]')
        title = title_heading.find_element(By.CLASS_NAME, "a-size-extra-large")
        title_data = title.text
        
        #타이틀 타입
        title_type = title_heading.find_element(By.ID, "title_type")
        title_type_data = title_type.text
        
        #러닝타임        
        running_time = title_heading.find_element(By.ID, "running_time")
        running_time_data = running_time.text
        
        #장르
        genres = title_heading.find_element(By.ID, "genres")
        genres_data = genres.text
        
    except:
        title_data = 'failed'
        title_type_data = 'failed'
        running_time_data = 'failed'
        genres_data = 'failed'
        
    try:
        meta_combined = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="title_heading"]/div')))
        meta_combined_data = meta_combined.text #데이터 줍기
    except:
        meta_combined_data = 'null'        
                
    ##tranding 영역
    try:
        tranding = wait.until(EC.presence_of_element_located((By.ID, 'popularity_widget')))
        moviemeter = tranding.find_element(By.CLASS_NAME, 'a-declarative')
        moviemeter_data= moviemeter.text
    except:
        moviemeter_data = 'failed'
        
    ##status 영역
    try:
        status_summary = wait.until(EC.presence_of_element_located((By.ID, 'status_summary')))
        released = status_summary.find_element(By.CLASS_NAME, 'a-declarative')
        released_data = released.text
    except:
        released_data = 'failed'
        

    driver.get(d)
    time.sleep(0.3)
    
    try:
        seemore = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="international_release_details"]/div/div[2]/a/span')))
        driver.execute_script("arguments[0].scrollIntoView();", seemore)
        #seemore를 click함
        driver.execute_script("arguments[0].click();", seemore)
        
    except:
        pass
    
    try:
        release_details = driver.find_element(By.XPATH, '//*[@id="international_release_details"]')
        table = release_details.find_element(By.CLASS_NAME, "a-keyvalue")
        
        target_column_index = 0  # 예시로 2번째 열을 대상으로 설정
        tbody = table.find_element(By.TAG_NAME, "tbody")
        rows = tbody.find_elements(By.TAG_NAME, "tr")
        
        # 특정 열의 모든 데이터를 담을 변수
        target_column_data_combined = ""
        
        for row in rows:
            # 행에서 특정 열(td)을 찾아내고 데이터를 변수에 추가
            columns = row.find_elements(By.TAG_NAME, "td")
            if len(columns) > target_column_index:
                target_column_data_combined += columns[target_column_index].text + ","
        
        # 마지막에 추가된 콤마 제거
        target_column_data_combined = target_column_data_combined.rstrip(", ")
        
        # 결과 출력 또는 원하는 작업 수행
    except:
        target_column_data_combined = 'failed'
        
    driver.get(c)
    driver.execute_script("window.scrollTo(0, 250);")
    time.sleep(0.3)
    
    try:
        distribution_wrapper = driver.find_element(By.ID, 'distribution_wrapper')
        table = distribution_wrapper.find_element(By.ID, "distribution")
        
        target_column_index = 0  # 예시로 2번째 열을 대상으로 설정
        tbody = table.find_element(By.TAG_NAME, "tbody")
        rows = tbody.find_elements(By.TAG_NAME, "tr")
        
        # 특정 열의 모든 데이터를 담을 변수
        distribition_combined = ""
        
        for row in rows:
            # 행에서 특정 열(td)을 찾아내고 데이터를 변수에 추가
            columns = row.find_elements(By.TAG_NAME, "td")
            if len(columns) > target_column_index:
                distribition_combined += columns[target_column_index].text + ","
        
        # 마지막에 추가된 콤마 제거
        distribition_combined = distribition_combined.rstrip(", ")
        
        # 결과 출력 또는 원하는 작업 수행
    except:
        distribition_combined = 'failed'
            
       
    data_dict = {
    'cast1': [cast1_data],
    'cast2': [cast2_data],
    'cast3': [cast3_data],
    'cast4': [cast4_data],
    'cast5': [cast5_data],
    'company': [company_data],
    'distributor': [distributor_data],
    'rating': [rating_data],
    'title': [title_data],
    'title_type': [title_type_data],
    'running_time': [running_time_data],
    'meta_combined': [meta_combined_data],
    'genres': [genres_data],
    'moviemeter Data': [moviemeter_data],
    'realease_date': [released_data],
    'release_country': [target_column_data_combined],
    'distribition': [distribition_combined],
    'url':[r]}
   
    page_df = pd.DataFrame(data_dict)
    
    # 결과 DataFrame에 현재 페이지의 데이터 추가
    result_df = pd.concat([result_df, page_df], ignore_index=True)
    
    driver.quit() 
     
print('셀레니움 동작완료')
```
